#2

- Média é a soma de todos os valores pela quantidade de valores
- Mediana é o valor ou a média dos 2 valores que estão nno centro dos valores ordenados
- Moda é o valor que mais se repete

Moda é ótima para trabalhar com distribuições e categorização
Média deve ser evitada quando houverem muitos outliers
Mediana é adeuada quando os dados estão ordenados

```python
#a (feita por IA)
cols = ["Idade", "Peso", "Temperatura", "N_Int"]
# Dicionário para guardar estatísticas
stats = {}
for col in cols:
    s = df[col]
    stats[col] = {
        "mean":     s.mean(),
        "median":   s.median(),
        "mode":     s.mode().tolist(),
        "std":      s.std(ddof=1),
        "var":      s.var(ddof=1),
        "Q1":       s.quantile(0.25),
        "Q3":       s.quantile(0.75),
        "skewness": s.skew(),
        "kurtosis": s.kurt()      # curtose de Pearson (3 = normal)
    }

# Converte para DataFrame e exibe
stats_df = pd.DataFrame(stats).T
stats_df
```

#b

```python
#b
quant_cols = ["Idade", "Peso", "Temperatura", "N_Int"]
sub = df[quant_cols]

# Matriz de Covariância
# cov(x,y)=E[(X-E[X])(Y-E[Y])]
# cov(X,Y)=E[XY]-E[X]E[Y]
cov_mat = sub.cov(ddof=1)
print("Matriz de Covariância:\n", cov_mat, "\n")

# Matriz de Correlação (Pearson)
# corr(x,y)=cov(X,Y)/sqrt(var(X)var(Y))
# corr(X,Y)=cov(X,Y)/sqrt(cov(X,X)cov(Y,Y))
# corr(X,Y)=cov(X,Y)/sqrt(var(X)var(Y))
corr_mat = sub.corr(method="pearson")
print("Matriz de Correlação:\n", corr_mat)
```
