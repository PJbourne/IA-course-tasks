```python
#3  A partir da regra de que oputliers têm valores menores que 1,5 * IQR e maiores que 1,5 * IQR, 
# identifique os atribnutos que possuem outliers e quais são esses valores.
Tq1 = df["Temperatura"].quantile(0.25)
Tq3 = df["Temperatura"].quantile(0.75)
Tiqr = Tq3 - Tq1
print(f"lim_inf temp: {Tq1 - 1.5 * Tiqr}")
print(f"lim_sup temp: {Tq3 + 1.5 * Tiqr}")
for i in df["Temperatura"]:
    if i < Tq1 - 1.5 * Tiqr or i > Tq3 + 1.5 * Tiqr:
        print(f"outlier temp: {i}")
Uq1 = df["Umidade"].quantile(0.25)
Uq3 = df["Umidade"].quantile(0.75)
Uiqr = Uq3 - Uq1
print(f"lim_inf umid: {Uq1 - 1.5 * Uiqr}")
print(f"lim_sup umid: {Uq3 + 1.5 * Uiqr}")
for i in df["Umidade"]:
    if i < Uq1 - 1.5 * Uiqr or i > Uq3 + 1.5 * Uiqr:
        print(f"outlier umid: {i}")
```

```python
#6 Usando a tabela com valores qualitativos (após a discretização pedida no exercício anterior),
# avalie a qualidade dos atributos usando a medida de ganho de informação 
# (ver seu cálculo no Capítulo 6).
# Como fica o ranqueamento dos atributos?
import pandas as pd
import numpy as np

# Função para calcular entropia
def entropia(coluna):
    probs = coluna.value_counts(normalize=True)
    return -sum(probs * np.log2(probs))

# Função para calcular ganho de informação
def ganho_informacao(df, atributo, classe):
    entropia_total = entropia(df[classe])
    valores = df[atributo].unique()
    
    entropia_condicional = 0
    for valor in valores:
        subset = df[df[atributo] == valor]
        peso = len(subset) / len(df)
        entropia_condicional += peso * entropia(subset[classe])
    
    return entropia_total - entropia_condicional

# Lista de atributos qualitativos para avaliar (após discretização)
atributos = ["Tempo", "Temperatura discretizada", "Umidade discretizada", "Vento"]
ganhos = {}

for atributo in atributos:
    ganhos[atributo] = ganho_informacao(df, atributo, "Joga")

# Mostrar o ranqueamento
ganhos_ordenados = sorted(ganhos.items(), key=lambda x: x[1], reverse=True)

print("Ganho de Informação por atributo:")
for atributo, ganho in ganhos_ordenados:
    print(f"{atributo}: {ganho:.4f}")
```
