```python
from sklearn.datasets import load_iris
```

```python
x, y = load_iris(return_X_y=True, as_frame=True)
```

```python
x = x.loc[y.isin([0,1]),['sepal length (cm)', 'sepal width (cm)']]
```

```python
y = y[y.isin([0,1])]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()
ax.scatter(x_train['sepal length (cm)'], x_train['sepal width (cm)'], c=y_train, cmap='coolwarm', label='Train')
plt.show()
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC(kernel='linear').fit(x_train, y_train)
```

```python
clf.coef_
```

```python
w1 = clf.coef_[0][0]
w1
```

```python
w2 = clf.coef_[0][1]
w2
```

```python
w0 = clf.intercept_[0]
w0
```

```python
fig, ax = plt.subplots()
ax.scatter(x_train['sepal length (cm)'], x_train['sepal width (cm)'], c=y_train, cmap='coolwarm', label='Train')

x = np.linspace(4, 7, 100)
y = (-w1 * x - w0) / w2
ax.plot(x, y, 'r')

plt.show()
```

```python
clf.support_vectors_
```

```python
fig, ax = plt.subplots()
ax.scatter(x_train['sepal length (cm)'], x_train['sepal width (cm)'], c=y_train, cmap='coolwarm', label='Train')

x = np.linspace(4, 7, 100)
y = (-w1 * x - w0) / w2
ax.plot(x, y, 'r')

ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], c='r')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x_train['sepal length (cm)'], x_train['sepal width (cm)'], c=y_train, cmap='coolwarm', label='Train')

x = np.linspace(4, 7, 100)
y = (-w1 * x - w0) / w2
ax.plot(x, y, 'r')
y2 = (+1-w1 * x - w0) / w2
ax.plot(x, y2, '--r')
y3 = (-1-w1 * x - w0) / w2
ax.plot(x, y3, '--r')

ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], c='r')

plt.show()
```
